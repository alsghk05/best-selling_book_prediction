```python
from transformers import BertTokenizer, BertModel
import torch

# KoBERT 모델과 토크나이저 불러오기
tokenizer = BertTokenizer.from_pretrained('monologg/kobert')
model = BertModel.from_pretrained('monologg/kobert')
```

```python
# '도서 이름' 임베딩
input_texts_name = final_dataset_5['도서 이름'].tolist()
inputs_name = tokenizer(input_texts_name, return_tensors='pt', padding=True, truncation=True)

with torch.no_grad():
    outputs_name = model(**inputs_name)
    
# 도서 이름의 임베딩 결과
name_embeddings = outputs_name.last_hidden_state[:, 0, :].numpy()

# '줄거리' 임베딩
input_texts_desc = final_dataset_5['줄거리'].tolist()
inputs_desc = tokenizer(input_texts_desc, return_tensors='pt', padding=True, truncation=True)

with torch.no_grad():
    outputs_desc = model(**inputs_desc)

# 줄거리의 임베딩 결과
desc_embeddings = outputs_desc.last_hidden_state[:, 0, :].numpy()

# 결과: 두 개의 임베딩을 각각 반환
print("도서 이름 임베딩:", name_embeddings)
print("줄거리 임베딩:", desc_embeddings)
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup

# '줄거리' 열 전처리 함수
def preprocess_text(text):
    if pd.isna(text):
        return text
    
    # HTML 태그 제거
    text = BeautifulSoup(text, "html.parser").get_text()

    return text

# '출거리' 열 전처리 적용
final_dataset_5['줄거리'] = final_dataset_5['줄거리'].apply(preprocess_text)
final_dataset_5['작가'] = final_dataset_5['작가'].apply(preprocess_text)


# 결과 출력
final_dataset_5['줄거리'].head()
```

```python
import pandas as pd
import re
from nltk.tokenize import sent_tokenize, word_tokenize

# 필요 없는 문자 제거 함수 (괄호 []는 제거하지 않음)
def remove_unwanted_chars(tokens):
    if isinstance(tokens, list):
        return [remove_unwanted_chars(token) for token in tokens]
    elif isinstance(tokens, str):
        return re.sub(r'[^\[\]가-힣a-zA-Z0-9\s]', '', tokens)
    else:
        return tokens

# 필요 없는 문자 제거 적용 (괄호 []는 제거하지 않음)
final_dataset_5['줄거리'] = final_dataset_5['줄거리'].apply(lambda x: remove_unwanted_chars(x))
final_dataset_5['장르'] = final_dataset_5['장르'].apply(lambda x: remove_unwanted_chars(x))

final_dataset_5
```

```python
import pandas as pd
import re
from nltk.tokenize import sent_tokenize, word_tokenize

# 불용어 파일 불러오기
with open('bulyong_a.txt', 'r', encoding='utf-8') as file:
    stop_words = set(file.read().splitlines())

# 불용어 제거 함수
def remove_stop_words(tokens):
    if isinstance(tokens, list):
        return [remove_stop_words(token) for token in tokens if token]
    elif isinstance(tokens, str):
        return ' '.join([word for word in tokens.split() if word not in stop_words])
    else:
        return tokens
```

```python
# 불용어 제거 적용
final_dataset_5['줄거리 토큰화'] = final_dataset_5['줄거리 토큰화'].apply(remove_stop_words)
final_dataset_5['도서 이름 토큰화'] = final_dataset_5['도서 이름 토큰화'].apply(remove_stop_words)

# 결과 출력
final_dataset_5
```

```python
# 특수 문자 제거 함수
def remove_special_chars(token_list):
    if isinstance(token_list, list):
        return [re.sub(r'[^\w\s]', '', token) for token in token_list]
    return token_list

# '형태소 토큰화_줄거리' 열에 함수 적용
final_dataset_5['줄거리 토큰화'] = final_dataset_5['줄거리 토큰화'].apply(remove_special_chars)
c
```

```python
# 연속된 콤마를 하나로 바꾸고, 콤마 뒤에 대괄호가 바로 오는 경우 대괄호로만 대체하고 콤마 뒤에 텍스트가 없는 경우 콤마 삭제
def replace_multiple_commas(text):
    if isinstance(text, list):
        return [replace_multiple_commas(subtext) for subtext in text if subtext]  # 값이 없는 요소 삭제
    if isinstance(text, str):
        text = re.sub(r',\s*,+', ',', text)
        text = re.sub(r',\s*\]', ']', text)
        text = re.sub(r',\s*$', '', text)  # 콤마 뒤에 텍스트가 없으면 콤마 삭제
    return text

final_dataset_5['줄거리 토큰화'] = final_dataset_5['줄거리 토큰화'].apply(replace_multiple_commas)
final_dataset_5['도서 이름 토큰화'] = final_dataset_5['도서 이름 토큰화'].apply(replace_multiple_commas)
final_dataset_5
```
