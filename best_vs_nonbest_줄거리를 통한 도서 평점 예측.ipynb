```python
import pandas as pd

# 데이터 정의
data = {
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'entire': [120169, 109221, 96835, 79926, 63962, 47406, 31421, 16001],
    'best': [2046, 1884, 1659, 1352, 1099, 834, 564, 318]
}

# 데이터프레임 생성
year_ratio = pd.DataFrame(data)

# 데이터프레임 출력
year_ratio
```

```python
# 'non-best' 열 생성
year_ratio['non_best'] = year_ratio['entire'] - year_ratio['best']

# 데이터프레임 출력
year_ratio
```

```python
# 'ratio' 열 생성 (non-best와 best의 비율)
year_ratio['ratio'] = year_ratio['non_best'] / year_ratio['best']

# 데이터프레임 출력
year_ratio
```

```python
# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 6))

# 전체 데이터 그래프
ax1.plot(year_ratio['year'], year_ratio['entire'], 'k-o', label='Entire', color='black', linewidth=2, markersize=8)
ax1.set_xlabel('Year')
ax1.set_ylabel('Entire', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# 범례 추가
fig.legend(loc='lower left', bbox_to_anchor=(0.1,0.9))

# 그래프 제목 및 레이아웃 설정
plt.title('Yearly Entire Data and Ratio')
fig.tight_layout()

# 그래프 출력
plt.show()
```

```python
import pandas as pd
import numpy as np
import ast
import seaborn as sns
import matplotlib.pyplot as plt

# 문자열로 저장된 벡터를 쉼표로 구분하고 리스트로 변환
def convert_to_list(vec_str):
    vec_str = vec_str.replace('[', '').replace(']', '')
    vec_list = [float(x) for x in vec_str.split()]
    return vec_list

for col in ['Word2Vec_임베딩', '형태소_장르_Word2Vec_임베딩', '도서_이름_Word2Vec_임베딩', '작가_Word2Vec_임베딩', '출판사_Word2Vec_임베딩']:
    Correlation3[col] = Correlation3[col].apply(convert_to_list)

# 벡터를 분해하여 열로 추가
for col in ['Word2Vec_임베딩', '형태소_장르_Word2Vec_임베딩', '도서_이름_Word2Vec_임베딩', '작가_Word2Vec_임베딩', '출판사_Word2Vec_임베딩']:
    vectors = pd.DataFrame(Correlation3[col].tolist(), index=Correlation3.index)
    vectors.columns = [f'{col}_{i}' for i in range(vectors.shape[1])]
    Correlation3 = pd.concat([Correlation3, vectors], axis=1)

# 분해된 후 원본 벡터 열 삭제
Correlation3.drop(columns=['Word2Vec_임베딩', '형태소_장르_Word2Vec_임베딩', '도서_이름_Word2Vec_임베딩', '작가_Word2Vec_임베딩', '출판사_Word2Vec_임베딩'], inplace=True)

# 상관관계 행렬 계산
correlation_matrix = Correlation3.corr()

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

plt.title('Correlation Heatmap of Embedding Features')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# 각 연도별로 데이터 프레임을 로드했다고 가정 (df_2017_non, df_2018_non 등)
df_list = [df_2017_non_, df_2018_non_, df_2019_non_, df_2020_non_, df_2021_non_, df_2022_non_, df_2023_non_]
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023]

# 시각화를 위한 설정
sns.set(style="whitegrid")

# 연도별로 데이터를 처리하는 루프
for df, year in zip(df_list, years):
    # 장르별로 그룹핑하고 베스트 여부에 따른 분포 확인
    genre_group = df.groupby(['장르', 'best']).size().unstack(fill_value=0)

    # 베스트, 비베스트 비율 계산
    genre_group['total'] = genre_group.sum(axis=1)
    genre_group['best_ratio'] = genre_group['Y'] / genre_group['total']
    genre_group['non_best_ratio'] = 1 - genre_group['best_ratio']

    # 시각화 (예: barplot)
    genre_group[['best_ratio', 'non_best_ratio']].plot(kind='bar', stacked=True)
    plt.title(f'Genre-wise Best vs Non-Best Ratio in {year}')
    plt.ylabel('Ratio')
    plt.xlabel('Genre')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
