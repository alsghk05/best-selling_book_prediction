```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# TF-IDF 벡터라이저 초기화
vectorizer = TfidfVectorizer()

# 'Okt_줄거리' 열의 내용을 문자열로 결합하여 TF-IDF 벡터화
documents = [' '.join(eval(row)) for row in sampled_df['Okt_줄거리']]

# TF-IDF 행렬 생성
tfidf_matrix = vectorizer.fit_transform(documents)

# 단어 목록 추출
feature_names = vectorizer.get_feature_names_out()

# 각 문서에서 가장 높은 TF-IDF 점수를 가진 단어 추출
def get_top_n_words(tfidf_vector, feature_names, top_n=5):
    # TF-IDF 점수와 단어의 인덱스를 매핑
    sorted_items = np.argsort(tfidf_vector.toarray()).flatten()[::-1]
    top_n_indices = sorted_items[:top_n]
    top_words = [feature_names[i] for i in top_n_indices]
    return top_words

# 주제 단어를 담을 열 추가
sampled_df['subject'] = [
    ', '.join(get_top_n_words(tfidf_matrix[i], feature_names, top_n=5)) 
    for i in range(tfidf_matrix.shape[0])
]

sampled_df
```

```python
import pandas as pd
from gensim.models import Word2Vec, FastText
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity

# GloVe 모델 로드 (미리 학습된 GloVe 모델)
glove_model = api.load("glove-wiki-gigaword-100")

# Word2Vec과 FastText 모델 학습 함수
def train_word2vec(sentences):
    model = Word2Vec(sentences, vector_size=100, window=5, sg=1, min_count=1)
    return model

def train_fasttext(sentences):
    model = FastText(sentences, vector_size=100, window=5, sg=1, min_count=1)
    return model

# 주제어와 다른 단어들 간의 코사인 유사도를 계산하여 상위 10개의 유사 단어를 추출하는 함수
def calculate_similarity(embedding_model, subject_words, all_words, top_n=10, model_type='word2vec'):
    similar_words_dict = {}
    
    for subject in subject_words:
        if model_type == 'glove':  # GloVe 모델의 경우
            if subject not in embedding_model:
                similar_words_dict[subject] = []
                continue
            subject_vector = embedding_model[subject].reshape(1, -1)
        else:  # Word2Vec 또는 FastText 모델의 경우
            if subject not in embedding_model.wv:
                similar_words_dict[subject] = []
                continue
            subject_vector = embedding_model.wv[subject].reshape(1, -1)
        
        similarities = {}
        for word in all_words:
            if (model_type == 'glove' and word in embedding_model) or (model_type != 'glove' and word in embedding_model.wv):
                if model_type == 'glove':
                    word_vector = embedding_model[word].reshape(1, -1)
                else:
                    word_vector = embedding_model.wv[word].reshape(1, -1)
                similarities[word] = cosine_similarity(subject_vector, word_vector)[0][0]
        
        sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
        similar_words_dict[subject] = [word for word, sim in sorted_similarities[:top_n]]
    
    return similar_words_dict

# 줄거리 데이터를 문장 형태로 변환
sentences = sampled_df['표제어_줄거리'].apply(lambda x: x).tolist()

# Word2Vec과 FastText 모델 학습
word2vec_model = train_word2vec(sentences)
fasttext_model = train_fasttext(sentences)

# 모든 단어들을 리스트로 정의 (유사도를 계산할 단어들)
all_words = list(set([word for description in sampled_df['표제어_줄거리'] for word in description]))

# 각 주제어에 대해 GloVe, Word2Vec, FastText로 유사 단어를 추출 (subject의 각 단어에 대해 개별적으로)
sampled_df['subject_words'] = sampled_df['subject'].apply(lambda x: x.split(', '))

sampled_df['GloVe_similar_words'] = sampled_df['subject_words'].apply(lambda x: calculate_similarity(glove_model, x, all_words, model_type='glove'))
sampled_df['Word2Vec_similar_words'] = sampled_df['subject_words'].apply(lambda x: calculate_similarity(word2vec_model, x, all_words, model_type='word2vec'))
sampled_df['FastText_similar_words'] = sampled_df['subject_words'].apply(lambda x: calculate_similarity(fasttext_model, x, all_words, model_type='fasttext'))

# 결과 출력
sampled_df[['subject', 'GloVe_similar_words', 'Word2Vec_similar_words', 'FastText_similar_words']]
```

```python
from kobert_transformers import get_tokenizer, get_kobert_model
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# KoBERT 모델 및 토크나이저 불러오기
model = get_kobert_model()
tokenizer = get_tokenizer()

# 단어 임베딩을 구하는 함수
def get_kobert_embedding(word):
    inputs = tokenizer(word, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# 주제어와 다른 단어들 간의 코사인 유사도를 계산하는 함수
def calculate_kobert_similarity(subject_words, all_words, top_n=10):
    similar_words_dict = {}
    
    for subject in subject_words:
        subject_vector = get_kobert_embedding(subject).reshape(1, -1)
        
        similarities = {}
        for word in all_words:
            word_vector = get_kobert_embedding(word).reshape(1, -1)
            similarities[word] = cosine_similarity(subject_vector, word_vector)[0][0]
        
        sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
        similar_words_dict[subject] = [word for word, sim in sorted_similarities[:top_n]]
    
    return similar_words_dict

# 모든 단어들을 리스트로 정의 (유사도를 계산할 단어들)
all_words = list(set([word for description in sampled_df_origin['줄거리'] for word in description.split()]))

# 각 주제어에 대해 KoBERT로 유사 단어 추출
sampled_df_origin['subject_words'] = sampled_df_origin['subject'].apply(lambda x: x.split(', '))

sampled_df_origin['KoBERT_similar_words'] = sampled_df_origin['subject_words'].apply(lambda x: calculate_kobert_similarity(x, all_words))

# 결과 출력
sampled_df_origin[['subject', 'KoBERT_similar_words']]
```

```python
import pandas as pd
from gensim.models import Word2Vec, FastText
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity

# GloVe 모델 로드 (미리 학습된 GloVe 모델)
glove_model = api.load("glove-wiki-gigaword-100")

# Word2Vec과 FastText 모델 학습 함수
def train_word2vec(sentences):
    model = Word2Vec(sentences, vector_size=100, window=5, sg=1, min_count=1)
    return model

def train_fasttext(sentences):
    model = FastText(sentences, vector_size=100, window=5, sg=1, min_count=1)
    return model

# 주제어와 다른 단어들 간의 코사인 유사도를 계산하여 상위 10개의 유사 단어 및 유사도를 추출하는 함수
def calculate_similarity(embedding_model, subject_words, all_words, top_n=10, model_type='word2vec'):
    similar_words_list = []
    
    for subject in subject_words:
        if model_type == 'glove':  # GloVe 모델의 경우
            if subject not in embedding_model:
                continue
            subject_vector = embedding_model[subject].reshape(1, -1)
        else:  # Word2Vec 또는 FastText 모델의 경우
            if subject not in embedding_model.wv:
                continue
            subject_vector = embedding_model.wv[subject].reshape(1, -1)
        
        similarities = {}
        for word in all_words:
            if (model_type == 'glove' and word in embedding_model) or (model_type != 'glove' and word in embedding_model.wv):
                if model_type == 'glove':
                    word_vector = embedding_model[word].reshape(1, -1)
                else:
                    word_vector = embedding_model.wv[word].reshape(1, -1)
                similarities[word] = cosine_similarity(subject_vector, word_vector)[0][0]
        
        # 유사도가 높은 순으로 정렬한 후, (단어, 유사도) 형식으로 저장
        sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
        for word, sim in sorted_similarities[:top_n]:
            similar_words_list.append({
                'subject': subject,
                'similar_word': word,
                'similarity': sim
            })
    
    return pd.DataFrame(similar_words_list)

# 줄거리 데이터를 문장 형태로 변환
sentences = sampled_df['표제어_줄거리'].apply(lambda x: x).tolist()

# Word2Vec과 FastText 모델 학습
word2vec_model = train_word2vec(sentences)
fasttext_model = train_fasttext(sentences)

# 모든 단어들을 리스트로 정의 (유사도를 계산할 단어들)
all_words = list(set([word for description in sampled_df['표제어_줄거리'] for word in description]))

# 각 주제어에 대해 GloVe, Word2Vec, FastText로 유사 단어를 추출 (subject의 각 단어에 대해 개별적으로)
sampled_df['subject_words'] = sampled_df['subject'].apply(lambda x: x.split(', '))

# GloVe 유사 단어 및 유사도 데이터프레임
glove_sim_df = pd.concat(sampled_df['subject_words'].apply(lambda x: calculate_similarity(glove_model, x, all_words, model_type='glove')).tolist())

# Word2Vec 유사 단어 및 유사도 데이터프레임
word2vec_sim_df = pd.concat(sampled_df['subject_words'].apply(lambda x: calculate_similarity(word2vec_model, x, all_words, model_type='word2vec')).tolist())

# FastText 유사 단어 및 유사도 데이터프레임
fasttext_sim_df = pd.concat(sampled_df['subject_words'].apply(lambda x: calculate_similarity(fasttext_model, x, all_words, model_type='fasttext')).tolist())

# 결과 출력
print("GloVe Similarity DataFrame:")
print(glove_sim_df)

print("\nWord2Vec Similarity DataFrame:")
print(word2vec_sim_df)

print("\nFastText Similarity DataFrame:")
print(fasttext_sim_df)
```

사전 지식 기반 군집 개수를 결정 및 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

# t-SNE 차원 축소 및 군집화 시각화 함수
def visualize_similarity(embedding_model, similar_words_dict, model_type='word2vec', n_clusters=5):
    # 유사 단어들의 벡터를 수집
    words = []
    vectors = []
    
    for subject, similar_words in similar_words_dict.items():
        for word in similar_words:
            if model_type == 'glove':
                if word in embedding_model:
                    vectors.append(embedding_model[word])
                    words.append(word)
            else:
                if word in embedding_model.wv:
                    vectors.append(embedding_model.wv[word])
                    words.append(word)
    
    # 벡터가 비어 있는 경우 방어 코드
    if not vectors:
        print(f"No vectors found for {model_type}.")
        return
    
    vectors = np.array(vectors)

    # t-SNE로 차원 축소 (2D)
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(vectors)

    # K-Means 군집화
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    y_kmeans = kmeans.fit_predict(X_tsne)

    # 시각화
    plt.figure(figsize=(14, 10))
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_kmeans, cmap='viridis', s=100)

    # 각 단어 레이블 추가
    for i, word in enumerate(words):
        plt.text(X_tsne[i, 0], X_tsne[i, 1], word, fontsize=9)

    plt.title(f't-SNE Visualization for {model_type.capitalize()} Model')
    plt.colorbar(label='Cluster')
    plt.show()

# 시각화 실행: GloVe, Word2Vec, FastText에 대해 각각 시각화
visualize_similarity(glove_model, sampled_df['GloVe_similar_words'].iloc[0], model_type='glove', n_clusters=5)
visualize_similarity(word2vec_model, sampled_df['Word2Vec_similar_words'].iloc[0], model_type='word2vec', n_clusters=5)
visualize_similarity(fasttext_model, sampled_df['FastText_similar_words'].iloc[0], model_type='fasttext', n_clusters=5)
```

엘보 방법으로 군집 수 결정 및 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

# 엘보 방법을 사용한 군집 수 결정 함수
def find_optimal_clusters(embedding_model, similar_words_dict, model_type='word2vec', max_clusters=10):
    words = []
    vectors = []
    
    # 유사 단어들의 벡터 수집
    for subject, similar_words in similar_words_dict.items():
        for word in similar_words:
            if model_type == 'glove':
                if word in embedding_model:
                    vectors.append(embedding_model[word])
                    words.append(word)
            else:
                if word in embedding_model.wv:
                    vectors.append(embedding_model.wv[word])
                    words.append(word)
    
    # 벡터가 비어 있는 경우 방어 코드
    if not vectors:
        print(f"No vectors found for {model_type}.")
        return
    
    vectors = np.array(vectors)

    # 엘보 방법을 통해 군집 수 결정
    sse = []
    cluster_range = range(1, max_clusters+1)
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(vectors)
        sse.append(kmeans.inertia_)  # inertia_는 SSE(WCSS) 값
    
    # 엘보 방법 시각화
    plt.figure(figsize=(8, 6))
    plt.plot(cluster_range, sse, marker='o')
    plt.title(f'Elbow Method for {model_type.capitalize()} Model')
    plt.xlabel('Number of clusters')
    plt.ylabel('SSE (Within-Cluster Sum of Squares)')
    plt.grid(True)
    plt.show()
```

```python
# GloVe, Word2Vec, FastText에 대해 엘보 방법 실행
find_optimal_clusters(glove_model, sampled_df['GloVe_similar_words'].iloc[2], model_type='glove', max_clusters=10)
find_optimal_clusters(word2vec_model, sampled_df['Word2Vec_similar_words'].iloc[2], model_type='word2vec', max_clusters=10)
find_optimal_clusters(fasttext_model, sampled_df['FastText_similar_words'].iloc[2], model_type='fasttext', max_clusters=10)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 유사도 그래프 생성 및 시각화 함수
def visualize_similarity_graph(embedding_model, subject, all_words, model_type='word2vec'):
    G = nx.Graph()

    # 주제어 단어들을 노드로 추가
    for word in subject:
        if model_type == 'glove' and word in embedding_model:
            G.add_node(word)
        elif model_type != 'glove' and word in embedding_model.wv:
            G.add_node(word)

    # 각 단어들 간 유사도를 계산하여 엣지를 추가
    for word1 in subject:
        for word2 in all_words:
            if word1 != word2:
                similarity = None  # 기본값 설정
                if model_type == 'glove':
                    if word1 in embedding_model and word2 in embedding_model:
                        subject_vector = np.array(embedding_model[word1]).reshape(1, -1)
                        word_vector = np.array(embedding_model[word2]).reshape(1, -1)
                        similarity = cosine_similarity(subject_vector, word_vector)[0][0]
                else:
                    if word1 in embedding_model.wv and word2 in embedding_model.wv:
                        subject_vector = np.array(embedding_model.wv[word1]).reshape(1, -1)
                        word_vector = np.array(embedding_model.wv[word2]).reshape(1, -1)
                        similarity = cosine_similarity(subject_vector, word_vector)[0][0]

                # 유사도가 0.5 이상일 때만 엣지를 추가 (기준은 조정 가능)
                if similarity is not None and similarity > 0.5:
                    G.add_edge(word1, word2, weight=similarity)

    # 그래프 레이아웃 설정
    pos = nx.spring_layout(G, k=0.3)

    # 엣지 가중치에 따라 엣지 두께 설정
    edges = G.edges(data=True)
    edge_weights = [G[u][v]['weight'] * 5 for u, v, d in edges]

    # 그래프 시각화
    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=3000, font_size=10, font_weight='bold', edge_color=edge_weights, width=edge_weights, edge_cmap=plt.cm.Blues)
    plt.title(f"Similarity Graph for {model_type.capitalize()} Model")
    plt.show()

# 시각화 실행
visualize_similarity_graph(glove_model, sampled_df['subject'].iloc[2].split(', '), all_words, model_type='glove')
visualize_similarity_graph(word2vec_model, sampled_df['subject'].iloc[2].split(', '), all_words, model_type='word2vec')
visualize_similarity_graph(fasttext_model, sampled_df['subject'].iloc[2].split(', '), all_words, model_type='fasttext')
```

```python
import pandas as pd
from kats.consts import TimeSeries
from kats.models import HoltWintersModel
import matplotlib.pyplot as plt

# 각 임베딩 모델의 유사도 점수 데이터프레임 생성 (예시)
# df는 subject, glove_sim, word2vec_sim, fasttext_sim 열을 포함해야 함
df = pd.DataFrame({
    'subject': ['example1', 'example2', 'example3'],
    'glove_sim': [0.8, 0.6, 0.9],
    'word2vec_sim': [0.7, 0.5, 0.8],
    'fasttext_sim': [0.9, 0.4, 0.85]
})

# KATS의 TimeSeries 객체 생성
ts_glove = TimeSeries(time=df.index, value=df['glove_sim'])
ts_word2vec = TimeSeries(time=df.index, value=df['word2vec_sim'])
ts_fasttext = TimeSeries(time=df.index, value=df['fasttext_sim'])

# Holt-Winters 모델을 사용하여 예측
model_glove = HoltWintersModel(ts_glove)
model_word2vec = HoltWintersModel(ts_word2vec)
model_fasttext = HoltWintersModel(ts_fasttext)

model_glove.fit()
model_word2vec.fit()
model_fasttext.fit()

# 예측 수행
forecast_glove = model_glove.predict(3)  # 3개 시점 예측
forecast_word2vec = model_word2vec.predict(3)
forecast_fasttext = model_fasttext.predict(3)

# 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(ts_glove.time, ts_glove.value, label='GloVe Similarity', color='blue')
plt.plot(ts_word2vec.time, ts_word2vec.value, label='Word2Vec Similarity', color='green')
plt.plot(ts_fasttext.time, ts_fasttext.value, label='FastText Similarity', color='red')
plt.legend()
plt.xlabel('Time Index')
plt.ylabel('Similarity Score')
plt.title('Similarity Scores over Time for Different Models')
plt.show()
```
